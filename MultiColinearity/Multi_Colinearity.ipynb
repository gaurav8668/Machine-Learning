```python
import pandas as pd
import statsmodels.api as sm
```

```python
df = pd.read_csv("Advertising.csv", index_col=0)
```

```python
df.head()
```

```python
# X = df[['TV', 'radio', 'newspaper']]
# y = df['sales']
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
X.head()
```

```python
X.corr()
```

```python
X.shape
```

```python
## fit a OLS moeel with intercept on TV and Radio
X = sm.add_constant(X)
print(type(X))
model = sm.OLS(y, X).fit()
```
